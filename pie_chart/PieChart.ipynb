```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_fuel_enigne_type = pd.read_csv('pie_chart_data.csv')
```

```python
df_fuel_enigne_type
```

```python
sns.set_palette('colorblind')
```

```python
plt.figure(figsize=(10,8))
plt.pie(df_fuel_enigne_type['Number of Cars'], labels = df_fuel_enigne_type['Engine Fuel Type'].values, autopct='%0.2f%%',textprops={'size':'x-large', 
                                                                                                                                     'fontweight':'bold',
                                                                                                                                    'color' : 'w'})
plt.title('Cars by Engine Fuel Type',fontsize = 18)
plt.legend()
plt.show()
```
