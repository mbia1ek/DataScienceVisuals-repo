```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
sns.set_style(style="white")
```

```python
df_hw = pd.read_csv("area_chart_homework_data.csv")
```

```python
df_hw
```

```python
colors = ['r','g','b']
labels = ['International','Services','Domestic']
plt.stackplot(df_hw['Date'].values,
              df_hw['International'].values,
              df_hw['Services'].values,
              df_hw['Domestic'].values,
              colors = colors,
              labels = labels)
plt.xticks(rotation = 90)
plt.show()
```
