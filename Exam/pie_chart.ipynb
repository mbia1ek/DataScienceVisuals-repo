```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv('pie_chart.csv')
```

```python
df
```

```python
plt.figure(figsize = (9,6))
plt.bar(x = df["Cities"], height=df["Sales"], color = "midnightblue")
plt.xticks(rotation = 45, fontsize = 13)
plt.yticks(fontsize = 13)
plt.title("Cars Listings by Brand", fontsize = 16, fontweight = 'bold')
plt.ylabel("Number of Listings", fontsize = 16)
plt.savefig("Used Cars Bar.png")
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.pie(df['Sales'], labels = df['Cities'].values, autopct='%0.2f%%',textprops={'size':'x-large', 
                                                                                                                                     'fontweight':'bold',
                                                                                                                                    'color' : 'w'})
plt.title('Cars by Engine Fuel Type',fontsize = 18)
plt.legend()
plt.show()
```

```python
plt.plot(df["Cities"], df["Cumulative Frequency"], color = "midnightblue")
plt.show()
```
