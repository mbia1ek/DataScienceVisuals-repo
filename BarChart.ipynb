```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_used_cars = pd.read_csv("bar_chart_data.csv")
```

```python
df_used_cars
```

```python
plt.figure(figsize = (9,6))
plt.bar(df_used_cars["Brand"], df_used_cars["Cars Listings"], color = "midnightblue")
plt.xticks(rotation = 45, fontsize = 13)
plt.yticks(fontsize = 13)
plt.title("Cars Listings by Brand", fontsize = 16, fontweight = 'bold')
plt.ylabel("Number of Listings", fontsize = 16)
plt.savefig("Used Cars Bar.png")
plt.show()
```

```python
plt.figure(figsize = (9,6))
plt.barh(df_used_cars["Brand"], df_used_cars["Cars Listings"], color = "midnightblue")
plt.yticks(rotation = 5, fontsize = 13)
plt.xticks(fontsize = 13)
plt.title("Cars Listings by Brand", fontsize = 16, fontweight = 'bold')
plt.xlabel("Number of Listings", fontsize = 16)
#plt.savefig("Used Cars Bar.png")
plt.show()
```
