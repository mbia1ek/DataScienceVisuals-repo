```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_ad_expenditure = pd.read_csv('scatter_plot_ii.csv')
```

```python
df_ad_expenditure
```

```python
#plt.figure(figsize=(10,8))
sns.set(rc={'figure.figsize':(9,6)})
sns.set_style(style='white')
sns.regplot(x='Budget',y='Sales',data=df_ad_expenditure, scatter_kws={'color':'k'},line_kws={'color':'red'})
plt.xlabel("Ad Expenditure in (000's $)")
plt.ylabel("Sales in (000's $)")
plt.title("Effect of Ad Expenditure on Sales", fontsize = 14, weight = 'bold', loc = 'left')
plt.show()
```

```python
import seaborn as sns
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt

sns.set_style(style='white')
data = sns.lmplot(x='Budget',y='Sales',data=df_ad_expenditure, scatter_kws={'color':'k'},line_kws={'color':'red'}, height=4, aspect=2)

def annotate(data, **kws):
    r, p = sp.stats.pearsonr(df_ad_expenditure['Budget'], df_ad_expenditure['Sales'])
    ax = plt.gca()
    ax.text(.05, .8, 'r={:.2f}, p={:.2g}'.format(r, p),
            transform=ax.transAxes)

data.map_dataframe(annotate)

plt.xlabel("Ad Expenditure in (000's $)")
plt.ylabel("Sales in (000's $)")
plt.title("Effect of Ad Expenditure on Sales", fontsize = 14, weight = 'bold', loc = 'left')
plt.show()
```
